```python
import pandas as pd

df = pd.read_csv("covid_faq.csv")
df.head()
```

```python
df.columns = ['question', 'answer']
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
device = 'cuda'

question_embeddings = model.encode(df['question'].tolist(),convert_to_tensor=True, device=device)

```

```python
def find_answer(user_query, top_k=1):
    query_embedding = model.encode(user_query, convert_to_tensor=True, device=device)
    similarities = util.pytorch_cos_sim(query_embedding, question_embeddings)[0]
    best_score, best_idx = similarities.topk(top_k)
    for score, idx in zip(best_score, best_idx):
        print(f"\n✅ Question: {df['question'][idx.item()]}")
        print(f"💬 Answer: {df['answer'][idx.item()]}")
        print(f"(Similarity: {score.item():.4f})")
```

```python
from sentence_transformers import util
find_answer("What is covid?")
```

```python
find_answer("What are the prevention from covid?")
```
